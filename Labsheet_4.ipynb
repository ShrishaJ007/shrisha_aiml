```python
import pandas as pd
```

```python
df = pd.read_csv('/content/German_Credit_Card_Dataset.csv')
df.info()
df.shape
df.head()
```

```python
df.iloc[0:5, 7:15]
```

```python
df['checkin_acc'].unique()
```

```python
X_features = list(df.columns)
X_features.remove('status')
encoded_df = pd.get_dummies(df[X_features], drop_first=True)
print(list(encoded_df.columns))
```

```python
X = encoded_df
y = df['status']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion = 'gini', max_depth = 3)
clf.fit(X_train, y_train)
```

```python
pred_y = clf.predict(X_test)
```

```python
from sklearn import metrics
print("Confusion Matrix is\n", metrics.confusion_matrix(pred_y, y_test))
print("Accuracy is\n", metrics.accuracy_score(y_test, pred_y))
print("AUC Score is", metrics.roc_auc_score(y_test, pred_y))
```

```python
from sklearn.tree import export_graphviz
import pydotplus as plot
from IPython.display import Image
export_graphviz(clf, out_file = "tree.odt", feature_names = X_train.columns, class_names = ['0', '1'], filled = True, rounded = True)
graph = plot.graphviz.graph_from_dot_file("tree.odt")
graph.write_png("tree.png")
Image(filename = "tree.png")
```
